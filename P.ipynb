```python
import numpy as np
from numpy import random as rand

n = 9
scenes= np.linspace(0, n-1, n, dtype=int)
S = rand.randint(10, 15, size=(n,n), dtype=int)
np.fill_diagonal(S, 0)
durations = rand.randint(10, 20, n)
W = 40
# np.ceil(sum(durations)/2 + rand.randint(1, 10))

W, durations, S
```

```python
increasingDurations = np.sort(durations)
increasingChangeover = np.array(np.sort(np.matrix.flatten(S)))

vals = []
hs = []
for h in range(2, n+1):
  total = np.sum(increasingChangeover[:h]) + np.sum(increasingDurations[:h])
  if total <= W:
    # to make sure combination is within admissible set
    vals.append(total)
    hs.append(h)
H = hs[np.argmax(vals)]
H
```

```python
setOfIdx = set([])
def finding_P(n, S, durations, W, H, total=0, idx = [], idxSet = set([])):
   
    for i in range(n):
        print("i: " + str(i))
        if i in idx:
            continue
        elif (len(idx) == H) and (i == n-1):
            continue

        if tuple(idx + [i]) in idxSet:
            continue
        
        if len(idx) >= H:
            continue
        elif not len(idx):
            temp_total = total + durations[i]
        else:
            temp_total = total + durations[i] + S[idx[-1], i]
        
        if (temp_total) <= W:
            temp_idx = idx + [i]
            
            idxSet.add(tuple(temp_idx))
            print("idx: " + str(temp_idx))
            finding_P(n, S, durations, W, H, temp_total, temp_idx, idxSet)
        
    print("idxSet: " + str(idxSet))
    print(f"ending {i} loop")
    return idxSet

finding_P(n, S, durations, W, H)
```
