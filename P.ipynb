```python
import numpy as np
from numpy import random as rand
import sys

sys.setrecursionlimit(10**6)
n = 9
scenes= np.linspace(0, n-1, n, dtype=int)
S = rand.randint(1, 10, size=(n,n), dtype=int)
np.fill_diagonal(S, 0)
durations = rand.randint(1, 10, n)
W = 50
# np.ceil(sum(durations)/2 + rand.randint(1, 10))

W, durations, S
```

```python
setOfIdx = set([])
def finding_P(n, S, durations, W, total=0, idx = [], idxSet = set([])):
   
    for i in range(n):
        print("i: " + str(i))
        if i in idx:
            continue
        elif len(idx) == n and i == n-1:
            continue
        
        if not len(idx):
            temp_total = total + durations[i]
        else:
            temp_total = total + durations[i] + S[idx[-1], i]
        
        if (temp_total) <= W:
            total = temp_total
            idx.append(i)
            if tuple(idx) in idxSet:
                continue
            idxSet.add(tuple(idx))
            print("idx: " + str(idx))
            
            finding_P(n, S, durations, W, temp_total, idx, idxSet)
        
        total = 0
        idx = []
    print("idxSet: " + str(idxSet))
    print(f"ending {i} loop")
    return idxSet

finding_P(n, S, durations, W)
```
